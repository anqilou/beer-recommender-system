```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
reviews_df=pd.read_csv("reviews.csv")
reviews_df=reviews_df.fillna("NA")
attributes = ['malty', 'balanced', 'complex']

count_vectorizer = CountVectorizer(stop_words = 'english')
```

```python
attributes = " ".join(attributes)

def similarity(string):
    documents = [attributes, string]
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
    return cosine_similarity(tfidf_matrix[0:1], tfidf_matrix)[0][1]
```

```python
reviews_df['base_similarity'] = reviews_df['product_review'].map(similarity)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize 
import nltk
nltk.download('punkt')
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score['compound']
```

```python
def parse_for_sentiment(review, keyword):
    sent_list = sent_tokenize(review)
    neighbors = ''
    for sent in sent_list:
        if keyword in sent.lower():
            lower_words = sent.lower().split()
            words = sent.split()
            indices = (i for i,word in enumerate(lower_words) if word==keyword)
            for ind in indices:
                neighbors += ' '.join(words[ind - 5 : ind + 6])
            neighbors += ' '
    if len(neighbors) > 0:
        return sentiment_analyzer_scores(neighbors)
    else:
        return np.nan
```

```python
reviews_df['balanced_sent'] = reviews_df['product_review'].apply(parse_for_sentiment, args = ('balanced',))
reviews_df['malty_sent'] = reviews_df['product_review'].apply(parse_for_sentiment, args = ('malty',))
reviews_df['complex_sent'] = reviews_df['product_review'].apply(parse_for_sentiment, args = ('complex',))
```

```python
sent_groupby = reviews_df.groupby(['product_name'])[['balanced_sent','malty_sent','complex_sent']].mean()
sent_groupby['avg_sent'] = sent_groupby.mean(axis = 1)
sent_groupby[['avg_base_similarity','avg_user_rating']] = reviews_df.groupby(['product_name'])[['base_similarity','user_rating']].mean()
sent_groupby['eval_score_base'] = sent_groupby['avg_base_similarity'] + sent_groupby['avg_sent']
top_3_base = sent_groupby.sort_values(by=['eval_score_base'], ascending = False).index.values[:3]
sent_groupby.sort_values(by=['eval_score_base'], ascending = False).head(3)
```

### Compute % of reviews that mentioned a preferred attribute

```python
def exist_attr(review, attr):
    if attr in review.lower():
        return 1
    else:
        return 0
    
attributes = ['malty', 'balanced', 'complex']
for attr in attributes:
    reviews_df[attr] = reviews_df['product_review'].apply(exist_attr, args=(attr,))

attr_groupby = reviews_df.groupby('product_name')[['malty','balanced','complex']]
    
attr_percent = attr_groupby.sum()/attr_groupby.count()*100
attr_percent[attr_percent.index.isin(top_3_base)]

```

Our recommendation stays the same even if we use word vectors instead of plain vanilla bag-of-words cosine similarity. But we did notice that the similarity scores increased significantly after we used word vectors. One possible explanation is that the exact attributes words we picked (i.e. malty, balanced, complex) did not show up in the reviews a lot of times (low cosine similarity scores), but there were plenty of words that have similar meanings to these attributes appeared in the reviews (high spaCy similarity scores). We think this is acceptable because these words with similar meanings will describe the characteristics of a beer the same way as the attribute words we selected. Since the products we recommend are the same using these two methods, we can not analyze the difference in % of reviews that mention a preferred attribute. But we do believe that the identical results further indicate it’s reasonable to recommend these products to customers who are looking for balanced, malty, and complex beer.
